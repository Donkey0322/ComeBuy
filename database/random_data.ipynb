```python
def query(q):
    try:
        conn = psycopg2.connect(
            host="localhost",
            database="comebuy_db",
            port='5432',
            user="dev",
            password="dev")
        cursor = conn.cursor()
        cursor.execute(q)
        cursor.close()
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print('err', error)
def query2(q,s):
    try:
        conn = psycopg2.connect(
            host="localhost",
            database="comebuy_db",
            port='5432',
            user="dev",
            password="dev")
        cursor = conn.cursor()
        cursor.execute(q,s)
        cursor.close()
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print('err', error)
def select(q):
    try:
        conn = psycopg2.connect(
            host="localhost",
            database="comebuy_db",
            port='5432',
            user="dev",
            password="dev")
        cursor = conn.cursor()
        cursor.execute(q)
        r = cursor.fetchall()
        cursor.close()
        conn.commit()
        return(r)
    except (Exception, psycopg2.DatabaseError) as error:
        print('err', error)
```

```python
# 把 store 存進 table 裡
for region in stores.keys():
    r = regions[region]
    for county in stores[region].keys():
        c = counties[county]
        for dist in stores[region][county].keys():
            d = districts[dist]
            for store in stores[region][county][dist]:
                sql = '''insert into stores(region, county, district, name)
                         values (%(re)s, %(co)s, %(di)s, %(st)s)
                      '''
                param = {'re':r, 'co':c, 'di':d, 'st':store}
                query2(sql, param)
```
